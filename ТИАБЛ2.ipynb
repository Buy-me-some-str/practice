```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
X = np.array(df.total_rooms).reshape(-1, 1)
model1 = LinearRegression().fit(X, df.total_bedrooms)
x = np.arange(0, max(df.total_rooms), 1)
y = x*model1.coef_+model1.intercept_
model2 = LinearRegression().fit(X, df.population)
y2 = x*model2.coef_+model2.intercept_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 7))
ax1.scatter(df.total_rooms, df.total_bedrooms, color="white",edgecolor="black",s=9)
ax1.plot(x, y, c="crimson")
ax1.set_axisbelow(True)
ax1.grid(linewidth=2,color="mistyrose")
ax2.plot(x, y2, c="crimson")
ax2.scatter(X, df.population, color="white",edgecolor="black",s=9)
ax2.set_axisbelow(True)
ax2.grid(linewidth=2,color="mistyrose")
```

```python
from sklearn.datasets import fetch_california_housing, fetch_openml
df = fetch_openml(name="mnist_784", as_frame=True)
```

```python
X = df.data[0:4000]
y = df.target[0:4000]
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler = MinMaxScaler()
X_p = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X = X_p
```

```python
from sklearn.manifold import TSNE
import time
tsne1 = TSNE(n_components=2, perplexity=15, random_state=123)
tsne2 = TSNE(n_components=2, perplexity=25, random_state=123)
tsne3 = TSNE(n_components=2, perplexity=35, random_state=123)
```

```python
time_start_tsne1 = time.time()
tsne1_feutures = tsne1.fit_transform(X_p)
time_end_tsne1 = time.time()
time_start_tsne2 = time.time()
tsne2_feutures = tsne2.fit_transform(X_p)
time_end_tsne2 = time.time()
time_start_tsne3 = time.time()
tsne3_feutures = tsne3.fit_transform(X_p)
time_end_tsne3 = time.time()
```

```python
data = pd.DataFrame({"x1":tsne1_feutures[:,0], "x2":tsne1_feutures[:,1], "y":str(y)})
```

```python
import seaborn as sns

sns.scatterplot(x=tsne1_feutures[:,0], y=tsne1_feutures[:,1], hue=y)
```

```python
time_end_tsne3-time_start_tsne3
```

```python
tsne3d = TSNE(n_components=3, perplexity=25, random_state=123)
tsne3d_feutures = tsne3d.fit_transform(X_p)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
x = tsne3d_feutures[:,0]
y = tsne3d_feutures[:,1]
z = tsne3d_feutures[:,2]
ax.scatter(x, y, z, c=y)
plt.show()
```

```python
#!pip install umap-learn
import umap

umap_model = umap.UMAP(n_components=2, n_neighbors=25, min_dist=0.1, random_state=42).fit(X_p)
umap_feutures = umap_model.transform(X_p)
```

```python
import matplotlib.pyplot as plt

import seaborn as sns

sns.scatterplot(x=umap_feutures[:,0], y=umap_feutures[:,1], hue=y)
```

```python
#!pip install umap-learn
import umap.plot

umap.plot.points(umap_model, labels=y, theme="darkblue")
```

```python
umap.plot.connectivity(umap_model, show_points=True)
```

```python
umap.plot.connectivity(umap_model, edge_bundling='hammer')
```

```python
n_n = (5, 25, 50)
m_d = (0.1, 0.6)
um = dict()
for i in range(len(n_n)):
    for j in range(len(m_d)):
        um[(n_n[i], m_d[j])]=(umap.UMAP(n_neighbors=n_n[i], min_dist=m_d[j], random_state=42).fit_transform(X_p))
```

```python
fig, ax = plt.subplots(3, 2, figsize=(15, 10))
cluster_colors = pd.DataFrame({"y":y}).groupby('y').ngroup()
for i in range(len(n_n)):
    for j in range(len(m_d)):
        ax[i][j].scatter(x=um.get((n_n[i], m_d[j]))[:,0],
                         y=um.get((n_n[i], m_d[j]))[:,1],
                         c=cluster_colors)
        ax[i][j].set_title(f"n_neighbors: {n_n[i]}, min_dist: {m_d[j]}")
```

```python
cluster_colors = pd.DataFrame({"y":y}).groupby('y').ngroup()

plt.scatter(x=um.get((n_n[0], m_d[0]))[:,0],
            y=um.get((n_n[0], m_d[0]))[:,1],
            c=cluster_colors)
```
