```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ages = np.random.normal(30, 20, 100000)
plt.hist(ages)
```

```python
def normal_dist(x, mu, s):
    return 1/(s*(2*np.pi)**0.5)*np.e**(-(x-mu)**2/(2*s**2))
```

```python
def bootstrap(data, num_samples, size_selection):
    bootstrap_means = np.zeros(num_samples)
    bootstrap_stds = np.zeros(num_samples)
    for i in range(num_samples):
        selection = np.random.choice(data,
                                     size=size_selection,
                                     replace=False)
        bootstrap_means[i] = selection.mean()
        bootstrap_stds[i] = selection.std()
    return bootstrap_means.mean(), bootstrap_stds.mean()
```

```python
mean, std = bootstrap(ages, 1000, 10)
```

```python
x = np.arange(min(ages), max(ages), 1)
y = normal_dist(x, mean, std)
```

```python
plt.hist(ages,
         bins=35,
         edgecolor = "black",
         density=True)
plt.axline((mean, 0),(mean, 0.02),
           color="red",
           linewidth=3)
plt.plot(x, y, color="red")
```
